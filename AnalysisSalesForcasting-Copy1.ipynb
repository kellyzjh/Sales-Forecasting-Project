```python
# --- Ensure sales target is defined ---
y = df['sales']

# --- Time features ---
df['year'] = df['order_date'].dt.year
df['month'] = df['order_date'].dt.month
df['quarter'] = df['order_date'].dt.quarter
df['weekday'] = df['order_date'].dt.day_name()

X_time = df[['year','month','quarter']]
X_weekday = pd.get_dummies(df['weekday'], drop_first=True)

# --- Categorical features ---
X_categorical = pd.get_dummies(df[['category','sub_category','segment','region','ship_mode']], drop_first=True)

# --- Lag features (shifted sales) ---
df['sales_lag_1'] = df['sales'].shift(1)
df['sales_lag_7'] = df['sales'].shift(7)
df['sales_lag_30'] = df['sales'].shift(30)
X_lag = df[['sales_lag_1','sales_lag_7','sales_lag_30']]

# --- Combine all predictors ---
X = pd.concat([X_time, X_weekday, X_categorical, X_lag], axis=1)

# Drop rows with NaN from lag features
X = X.dropna()

# Align target variable with X
y_final = y[X.index]

# Quick check
print("Predictor shape:", X.shape)
print("Target shape:", y_final.shape)
```

```python
# --- IMPORT MODELS AND METRICS ---
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# --- SPLIT DATA INTO TRAIN AND TEST ---
X_train, X_test, y_train, y_test = train_test_split(
    X, y_final, test_size=0.2, random_state=42
)

# --- INITIALIZE AND TRAIN RANDOM FOREST REGRESSOR ---
rf_model = RandomForestRegressor(
    n_estimators=200,        # number of trees
    max_depth=10,            # maximum depth of each tree
    random_state=42,
    n_jobs=-1                # use all processors
)
rf_model.fit(X_train, y_train)

# --- MAKE PREDICTIONS ---
y_pred = rf_model.predict(X_test)

# --- EVALUATE MODEL ---
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print(f"Random Forest Performance:")
print(f"RMSE: {rmse:.2f}")
print(f"R²: {r2:.2f}")

# --- FEATURE IMPORTANCE ---
importances = pd.Series(rf_model.feature_importances_, index=X.columns)
top_features = importances.sort_values(ascending=False).head(15)

# Plot top 15 important features
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
top_features.plot(kind='bar', color='skyblue')
plt.title("Top 15 Features Affecting Sales")
plt.ylabel("Importance")
plt.show()
```

```python
import pandas as pd
import numpy as np
from datetime import timedelta
from sklearn.ensemble import RandomForestRegressor
import matplotlib.pyplot as plt

# --- Minimal test dataset: last 60 days of daily sales ---
np.random.seed(42)
dates = pd.date_range(start='2026-01-01', periods=60)
sales = np.random.randint(1000, 5000, size=len(dates))
df_test = pd.DataFrame({'order_date': dates, 'sales': sales})
daily_sales = df_test.set_index('order_date')['sales'].sort_index()

# --- Create lag features for training ---
df_test['sales_lag_1'] = df_test['sales'].shift(1)
df_test['sales_lag_7'] = df_test['sales'].shift(7)
df_test = df_test.dropna()

# --- Train Random Forest on lag features only (minimal example) ---
X_train = df_test[['sales_lag_1', 'sales_lag_7']]
y_train = df_test['sales']

rf_model = RandomForestRegressor(n_estimators=50, random_state=42)
rf_model.fit(X_train, y_train)

# --- Forecast next 10 days ---
sales_extended = daily_sales.copy()
forecast_horizon = 10
forecast_dates = pd.date_range(start=sales_extended.index.max() + timedelta(days=1), periods=forecast_horizon)
forecast_df = pd.DataFrame(index=forecast_dates, columns=['forecast_sales'])

for i, current_date in enumerate(forecast_dates):
    lag_1 = sales_extended.get(current_date - timedelta(days=1), daily_sales.mean())
    lag_7 = sales_extended.get(current_date - timedelta(days=7), daily_sales.mean())
    
    X_forecast = pd.DataFrame([[lag_1, lag_7]], columns=['sales_lag_1','sales_lag_7'])
    predicted_sales = rf_model.predict(X_forecast)[0]
    
    forecast_df.loc[current_date, 'forecast_sales'] = predicted_sales
    sales_extended.loc[current_date] = predicted_sales
    
    print(f"Forecast day {i+1}: {current_date.date()} => {predicted_sales:.2f}")

# Plot results
plt.figure(figsize=(10,5))
plt.plot(daily_sales[-30:], label='Historical Sales')
plt.plot(forecast_df['forecast_sales'], label='Forecast', linestyle='--', color='orange')
plt.title("10-Day Sales Forecast (Test Example)")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
from datetime import timedelta
from sklearn.ensemble import RandomForestRegressor
import matplotlib.pyplot as plt

# --- Minimal test dataset (simulate last 30 days of sales) ---
np.random.seed(42)
dates = pd.date_range(start='2026-01-01', periods=30)
sales = np.random.randint(1000, 5000, size=len(dates))
df_test = pd.DataFrame({'order_date': dates, 'sales': sales})

# --- Create lag features for training ---
df_test['sales_lag_1'] = df_test['sales'].shift(1)
df_test['sales_lag_2'] = df_test['sales'].shift(2)
df_train = df_test.dropna()

# --- Define training data ---
X_train = df_train[['sales_lag_1', 'sales_lag_2']]
y_train = df_train['sales']

# --- Train Random Forest ---
rf_model = RandomForestRegressor(n_estimators=50, random_state=42)
rf_model.fit(X_train, y_train)

# --- Prepare for forecast ---
sales_extended = df_test.set_index('order_date')['sales'].sort_index().copy()
forecast_horizon = 5
forecast_dates = pd.date_range(start=sales_extended.index.max() + timedelta(days=1), periods=forecast_horizon)
forecast_df = pd.DataFrame(index=forecast_dates, columns=['forecast_sales'])

# --- Forecast loop ---
for i, current_date in enumerate(forecast_dates):
    lag_1 = sales_extended.get(current_date - timedelta(days=1), sales_extended.mean())
    lag_2 = sales_extended.get(current_date - timedelta(days=2), sales_extended.mean())

    X_forecast = pd.DataFrame([[lag_1, lag_2]], columns=['sales_lag_1','sales_lag_2'])
    predicted_sales = rf_model.predict(X_forecast)[0]
    
    forecast_df.loc[current_date, 'forecast_sales'] = predicted_sales
    sales_extended.loc[current_date] = predicted_sales

    print(f"Forecast day {i+1}: {current_date.date()} => {predicted_sales:.2f}")

# --- Plot results ---
plt.figure(figsize=(10,5))
plt.plot(sales_extended[-30:], label='Historical Sales')
plt.plot(forecast_df['forecast_sales'], label='Forecast', linestyle='--', color='orange')
plt.title("5-Day Sales Forecast (Minimal Test)")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.legend()
plt.show()
```
